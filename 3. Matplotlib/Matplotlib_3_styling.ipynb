```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,11)
```

```python
y = 2*x
```

```python
a = np.arange(0,10)
```

```python
b = a**4
```

```python
x,y,a,b
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,10),dpi=150)

axes[0][1].plot(x,y,"r")
axes[0][1].set_xlabel("Linear X")
axes[0][1].set_ylabel("Linear Y")
axes[0][1].set_title("Linear")

axes[1][0].plot(a,b,"g")
axes[1][0].set_xlabel("Exponential X")
axes[1][0].set_ylabel("Exponential Y")
axes[1][0].set_title("Exponential")
axes[1][0].plot(a, a**2,label="a**2")
axes[1][0].legend()

axes[0][0].plot(y,x/2)
axes[0][0].set_title("Some Relation")

axes[1][1].plot(b,a,"c")
axes[1][1].set_title("Inverted")



fig.suptitle("2x2 Linear and Exponential")

# fig.subplots_adjust(wspace=0.5,hspace=0.5) # manual adjusting

```

## setting color, line width and linetypes

```python
# MATLAB style line color and style. Not recommended, instead use named key word argument methods
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

```python
# key word argument method
fig, ax = plt.subplots()
ax.plot(x, x+1, color="blue", alpha=0.5)  # alpha to set transparency
ax.plot(x, x+2, color="#8B008B")  # rgb hex color codes
ax.plot(x, x+3, color= "#FF8C00")
```

### Line and marker style

### line width

```python
fig, ax = plt.subplots(figsize=(12,6))

# use linewidth or lw

ax.plot(x, x-1, color="red", linewidth=0.25)
ax.plot(x,x-2, color="red", lw=0.5)
ax.plot(x,x-3, color="red", lw=1)
ax.plot(x,x-4, color="red", lw=10)
```

### Line styles


```python
fig, ax = plt.subplots(figsize=(12,6))

# use linewidth or lw 
# possible linestype options ‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’

ax.plot(x, x-1, color="green", linewidth=3,linestyle="-") #solid dash
ax.plot(x,x-2, color="black", lw=3, ls="-.") # dot and dash
ax.plot(x,x-3, color="red", lw=3,ls=":") # dots
ax.plot(x,x-4, color="blue", lw=3, ls="--") # dash
```

### custom lines
#### set_dashes() method takes line length and space length combo

```python
fig, ax = plt.subplots(figsize=(12,6))

lines = ax.plot(x, x+10,color="green", lw=3)

lines[0].set_dashes([10,10])
```

### note that variable lines above is a list. we must index specific items to call set dashes method

```python
fig, ax = plt.subplots(figsize=(12,6))

lines = ax.plot(x, x+10,color="green", lw=3)

lines[0].set_dashes([5,2,5,2,10,3])  # a sequence in the form [line lenght,space lenght, line length, space length]
```

```python
fig, ax = plt.subplots(figsize=(12,6))

lines = ax.plot(x, x+10,color="green", lw=3)

lines[0].set_dashes([1,1,1,1,10,10]) 
```

### Markers
#### We've technically always been plotting points, and matplotlib has been automatically drawing a line between these points for us. We can place markers at each of these points.

#### There are several Markers Styles

```python
fig, ax = plt.subplots(figsize=(12,6))

# refer documentation for marker string codes
ax.plot(x, x+2, lw=3, marker="+", markersize=20)
ax.plot(x, x+5,lw=3, marker="o", ms=20)
ax.plot(x, x+7,lw=0, marker="s", ms=20) # linewidth zero shows only markers
ax.plot(x, x+9,lw=3, marker="1",ms=20)
```

### custom marker edges, thickness, size and style


```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x, color="cyan", lw=3, ls="-", marker="s", ms=20, markerfacecolor="red",
       markeredgewidth=8, markeredgecolor="green")
```
