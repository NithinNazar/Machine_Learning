```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6), dpi =200)
sns.scatterplot(x="salary", y="sales", data=df)
```

#### seaborn scatter plots takes x, y and data frame as main arguments and all other arguments are styling and customization

#### hue = sets the color points based on categorical data from data frame
#### palette = color palette for styling, must choose sequential or categorical palettes correctly from matplotlib 

```python
plt.figure(figsize=(12,6), dpi =200)
sns.scatterplot(x="salary", y="sales", data=df, hue="division")
```

```python
plt.figure(figsize=(12,6), dpi =200)
sns.scatterplot(x="salary", y="sales", data=df, hue = "work experience")  # providing a continuous data
```

```python
plt.figure(figsize=(12,6), dpi =200)
sns.scatterplot(x="salary", y="sales", data=df, hue="work experience",palette="viridis")
```

## scatterplot parameters :
#### These are more specific to scatterplot() call


## Size:
#### sets the size based on another column

```python
plt.figure(figsize=(12,6), dpi =200)
sns.scatterplot(x="salary", y="sales", data=df, size="work experience")
```

#### use s= if you want to change the size of the marker points to an integer value

```python
plt.figure(figsize=(12,6), dpi =200)
sns.scatterplot(x="salary", y="sales", data=df, s=200)
```

```python
plt.figure(figsize=(12,6), dpi =200)
sns.scatterplot(x="salary", y="sales", data=df,s=200, alpha=0.2, linewidth =0)
```

#### Style: automatically choses styles based on another categorical value from the dataframe
#### Optionally use the markers= parameter to pass a list of marker choices based off matplotlib, for example: ['*','+','o']

```python
plt.figure(figsize=(12,6), dpi =200)
sns.scatterplot(x="salary", y="sales", data=df, style="level of education")
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,style='level of education', hue="level of education",s=100)
```

### Exporting a seaborn figure

```python
plt.figure(figsize=(12,8), dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,style='level of education', hue="level of education", s=100)

plt.savefig("my_seaborn_plot_1.jpg")
```
