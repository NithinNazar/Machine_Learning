```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y = 2 * x
```

```python
y
```

## Matplotlib can plot functional relation or just data relations
## Matplotlib has basic method calls and oop method calls
## Matplotlib allows highlevel of customization of ploting and visualization

```python
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Graph Title Here")
plt.show()  # if you're on a non jupyter editor, but removes out cell info
```

### Editing more parameters

```python
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xlim(0,6) # upper limit, lower limit. cuts of the graph but actual data may present further beyond
plt.ylim(0,12)
plt.title("Graph Title Here")
plt.show()
```
