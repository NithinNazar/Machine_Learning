## Creating Figure
### a figure object is created and then all methods and attributes are callen on that object, this enables more functionality and flexibility, it is just like creating a blank canvas and then setting it up for plotting

```python
# creating Figure(empty canvas)
fig = plt.figure()
```

### NOTE ALL COMMANDS MUST GO IN SAME SELL
#### To begin with first create a figure instance to create a blank canvas, then we can add axes to that figure:

```python
# create an empty figure(empty canvas)
fig = plt.figure()

# add set of axes to the figure
axes = fig.add_axes([0,0,1,1])  # left, bottom, width, height (range 0 to 1)

# plot on that set of axes
axes.plot(a,b)
```

```python
# creating empty figure
fig = plt.figure()

# adding axes
axes = fig.add_axes([0,0,1,1])

# ploting over that axes
axes.plot(x,y)
```

### we can add multiple axes on the canvas and then plot on that axes


```python
type(fig)
```

```python
# creating empty figure
fig = plt.figure()

# adding axes 
axes_1 = fig.add_axes([0,0,1,1])  # large figure
axes_2 = fig.add_axes([0.2,0.2,0.4,0.4])  # smaller figure

# plotting larger figure on axes_1
axes_1.plot(a,b)

# adding labels

axes_1.set_xlabel("X Axis 1")
axes_1.set_ylabel("Y Axis 1")
axes_1.set_title("Larger Figure")

# inserting smaller figure into larger figure

axes_2.plot(a,b)
axes_2.set_title("Small figure")



```

### note that the axes of smaller figure is set proportionally to the size of the canvas
### 20 percent from left 20 from bottom 40 percent width and 40 height with respect to axes of larger figure which in default takes up entire size of canvas

#### further editing smaller figure

```python
# creating empty figure
fig = plt.figure()

# adding axes 
axes_1 = fig.add_axes([0,0,1,1])  # large figure
axes_2 = fig.add_axes([0.2,0.2,0.4,0.4])  # smaller figure

# plotting larger figure on axes_1
axes_1.plot(a,b)

# adding labels

axes_1.set_xlabel("X Axis 1")
axes_1.set_ylabel("Y Axis 1")
axes_1.set_title("Larger Figure")

# inserting smaller figure into larger figure

axes_2.plot(a,b)
axes_2.set_xlabel("x")
axes_2.set_ylabel("y")
axes_2.set_xlim(8,10)
axes_2.set_ylim(4000,10000)
axes_2.set_title("Zoomed In")
```

#### You can add as many axes on to the same figure as you want, even outside of the main figure if the length and width correspond to this.


```python
# creating empty figure
fig = plt.figure()

# adding axes 
axes_1 = fig.add_axes([0,0,1,1])  # large figure
axes_2 = fig.add_axes([0.2,0.2,0.4,0.4])  # smaller figure
axes_3 = fig.add_axes([1,1,0.25,0.25])

# plotting larger figure on axes_1
axes_1.plot(a,b)

# adding labels

axes_1.set_xlabel("X Axis 1")
axes_1.set_ylabel("Y Axis 1")
axes_1.set_title("Larger Figure")

# inserting smaller figure into larger figure

axes_2.plot(a,b)
axes_2.set_xlabel("x")
axes_2.set_ylabel("y")
axes_2.set_xlim(8,10)
axes_2.set_ylim(4000,10000)
axes_2.set_title("Zoomed In")

# another axes
axes_3.plot(a,b)
```

## Figure Parameters

```python
fig = plt.figure(figsize=(12,8), dpi=100)  # fig size is in inches

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)
```

## Exporting a figure

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(a,b)
axes.set_xlabel("X")
axes.set_ylabel("Y")
fig.savefig("figure_2.png",bbox_inches='tight')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])

axes1.plot(a,b)
axes1.set_xlabel('X')

# bbox_inches ='tight' automatically makes sure the bounding box is correct
fig.savefig('figure3.png',bbox_inches='tight')
```
