## plt.subplots()
### The plt.subplots() object will act as a more automatic axis manager. This makes it much easier to show multiple plots side by side. A tuple is used to grab the figure object and axes as np arrays


```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# now use axes object to add stuff to your plot
axes.plot(x,y,"g")  # color is optional. check documentation for color codes
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("SUBPLOTS")
```

## Plotting on axes objects
### Just as before, we simple .plot() on the axes objects, and we can also use the .set_ methods on each axes.

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for axis in axes:
    axis.plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(a,b,"r")
axes[0][1].plot(x,y,"g")
axes[1][0].plot(x,y,"b")
axes[1][1].plot(a,b,"c")

plt.tight_layout()  # adjusts the position of plots so that there will nbe no overlapping. fig.tight_layout() also works
```

## Parameters on subplots()
### Recall we have both the Figure object and the axes. Meaning we can edit properties at both levels.

```python
fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(12,8))  # tweak figsize to adjust the plot spacing and appearance

# SET YOUR AXES PARAMETES FIRST
axes[0][0].plot(a,b,"r")
axes[0][0].set_title("00 Red")

axes[0][1].plot(x,y,"b")
axes[0][1].set_title("01 Blue")

axes[1][0].plot(a,b,"g")
axes[1][0].set_title("10 Green")

axes[1][1].plot(x,y,"c")
axes[1][1].set_title("11 Cyan")
axes[1][1].set_xlabel("cx")
axes[1][1].set_ylabel("cy")

# THEN SET OVERALL FIGURE PARAMETERS
fig.suptitle("Colored Plots", fontsize=16)
```

### manual spacing on subplots()

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

# Parameters at the axes level
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

# Use left,right,top, bottom to stretch subplots
# Use wspace,hspace to add spacing between subplots

fig.subplots_adjust(left=None, bottom=None,right=None,top=None, wspace=1, hspace=1) # w and h space is fraction relative to h&w
```
